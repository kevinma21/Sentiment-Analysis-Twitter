```python
# Crawl Data

filename = 'dataset_kurikulum_merdeka2.csv'
search_keyword = 'kurikulum merdeka lang:id -is:reply -is:retweet'
limit = 2500
```

```python

!npx -y tweet-harvest@2.6.1 -o "{filename}" -s "{search_keyword}" --tab "LATEST" -l {limit} --token {twitter_auth_token}
```

```python
import pandas as pd

file_path = f"tweets-data/{filename}"

df = pd.read_csv(file_path, delimiter=",")

df_filtered = df[['conversation_id_str', 'created_at', 'full_text', 'username']]

output_file_path = f"tweets-data/filtered_{filename}"

df_filtered.to_csv(output_file_path, index=False)
```
